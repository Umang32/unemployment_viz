```python
import pandas as pd 
```

```python
df_2020 = pd.read_csv("Data/UE Benefits Search vs UE Rate 2004-20.csv")
```

```python
df_2019 = pd.read_csv("Data/UE Benefits Search vs UE Rate 2004-19.csv")
```

```python
df_2020.MONTH = pd.to_datetime(df_2020.MONTH)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

years = mdates.YearLocator()
years_format = mdates.DateFormatter("%Y")
months = mdates.MonthLocator()
```

```python
df_2020.dtypes
```

```python

plt.figure(figsize=(10,6), dpi=120) 
plt.title('"Unemployment Benefits"', fontsize=18)
plt.xticks(fontsize=13, rotation=45)
ax1 = plt.gca()
ax2 = ax1.twinx()


plt.xticks(fontsize=14, rotation=45)
ax1.grid(color='grey', linestyle='--')


ax1.set_xlim([df_2020.MONTH.min(), df_2020.MONTH.max()])
ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_format)
ax1.xaxis.set_minor_locator(months)



ax1.set_ylabel('UE_BENEFITS_WEB_SEARCH', color='#E6232E', fontsize=14)
ax2.set_ylabel('UNRATE', color='green', fontsize=14)


ax1.plot(df_2020.MONTH, df_2020.UNRATE , 'r', linestyle='dashed')
ax2.plot(df_2020.MONTH, df_2020.UE_BENEFITS_WEB_SEARCH, 'green')
```
